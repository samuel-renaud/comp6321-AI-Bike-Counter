```python
import pandas as pd
import numpy as np
```

```python
bike_count_2014 = pd.read_csv('bike_count/comptagevelo2014.csv')
bike_count_2014 = bike_count_2014.fillna(0)
bike_count_2014 = bike_count_2014[['Berri1','Maisonneuve_2','Maisonneuve_3']].astype('int64')

fields = ['Max Temp (°C)', 
          'Min Temp (°C)',
         'Mean Temp (°C)',
         'Total Precip (mm)',
          'Snow on Grnd (cm)',
         'Year',
         'Month',
         'Day']

weather_2014 = pd.read_csv('weather_daily/en_climate_daily_QC_702S006_2014_P1D.csv', usecols=fields)
weather_2014 = weather_2014.fillna(0)

fields = ['Dew Point Temp (°C)', 
          'Rel Hum (%)',
          'Wind Dir (10s deg)',
          'Wind Spd (km/h)',
         'Wind Chill']

weather_hourly_2014 = pd.read_csv('weather_hourly/en_climate_hourly_QC_702S006_01-2014_P1H.csv', usecols=fields)

for i in range(11):
    n = i+2
    name = open("weather_hourly/en_climate_hourly_QC_702S006_%.2d-2014_P1H.csv" % n)
    weather_hourly_2014_temp = pd.read_csv(name)
    weather_hourly_2014_temp
    weather_hourly_2014 = pd.concat([weather_hourly_2014, weather_hourly_2014_temp], axis=0, join="inner")

weather_hourly_2014 = weather_hourly_2014.fillna(0)
weather_hourly_2014 = weather_hourly_2014.groupby(np.arange(len(weather_hourly_2014))//24).mean()
weather_hourly_2014.reindex()

totalData_2014 = pd.concat([weather_2014,weather_hourly_2014,bike_count_2014], axis=1, join="inner")
totalData_2014.to_csv('totalData_2014.csv',index=False)
totalData_2014.info()
```
